```python
from matplotlib import pyplot as plt
```

```python
import numpy
```

```python
def fiterator(seed):
    x = seed
    orbit = []
    for i in range(4):
        x =  x*x
        orbit.append(x)
    return orbit
```

```python
o1 = fiterator(1.1)
o2 = fiterator(1.2)
o3 = fiterator(0.9)
o4 = fiterator(0.8)
t = [0,1,2,3]
```

```python
plt.plot(t,o1,label="Orbit1",linestyle='-',marker='o')
plt.plot(t,o2,label="Orbit2",linestyle='--',marker=',')
plt.plot(t,o3,label="Orbit3",linestyle='-.',marker='s')
plt.plot(t,o4,label="Orbit4",linestyle=':',marker='x')
plt.legend()
plt.show()
```
